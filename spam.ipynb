# load data set

```python
import pandas as pd

data_set = pd.read_csv("./emails.csv")
data_set = data_set.iloc[:,1:]
data_set.head(10)
```

```python
words = data_set.iloc[:,:3000]
perdiction = data_set.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

train_w, test_w, train_l, test_l = train_test_split(
    words, perdiction, test_size= 0.2, random_state= 42)

from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier()

mlp.fit(train_w, train_l)

perdiction_l = mlp.predict(test_w)

print("Preciision: %.5f" %precision_score(test_l, perdiction_l))
print("Recall: %.5f" %recall_score(test_l, perdiction_l))
print("F1 score: %.5f" %f1_score(test_l, perdiction_l))
print()

confusion_dp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(test_l, perdiction_l))
confusion_dp.plot()
```
